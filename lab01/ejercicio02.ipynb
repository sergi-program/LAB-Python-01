Punto 1. Condicionales para decidir si una persona puede conducir.

```python
edad = int(input("Ingrese su edad: "))
tiene_licencia = input("¿Tiene licencia? (s/n): ").strip().lower()
sancionado = input("¿Está sancionado? (s/n): ").strip().lower()

if edad >= 18 and tiene_licencia == "s" and sancionado == "n":
    print("Puede conducir")
elif edad < 18:
    print("No puede conducir porque es menor de edad")
elif tiene_licencia != "s":
    print("No puede conducir porque no tiene licencia")
else:
    print("No puede conducir porque está sancionado")
```

Punto 2. Ciclo for, suma de números naturales, números pares y temperaturas mayores al promedio.

```python
n = int(input("Ingrese un número n: "))

suma_naturales = 0
numeros_pares = []

for i in range(1, n + 1):
    suma_naturales += i
    if i % 2 == 0:
        numeros_pares.append(i)

print("Suma de los primeros números naturales:", suma_naturales)
print("Números pares encontrados:", numeros_pares)

temperaturas = [22.5, 23.0, 21.8, 24.3, 25.1]
promedio = sum(temperaturas) / len(temperaturas)
mayores_al_promedio = []

for temp in temperaturas:
    if temp > promedio:
        mayores_al_promedio.append(temp)

print("Promedio de temperaturas:", promedio)
print("Temperaturas mayores al promedio:", mayores_al_promedio)
```

Punto 3. Ciclo while, conteo hacia atrás y lectura de números hasta que el usuario ingrese 0.

El estado del ciclo está representado por la variable contador en el primer caso y por la variable acumulada suma_entradas junto con el valor ingresado en el segundo caso.

```python
contador = 5
while contador >= 0:
    print("Contador:", contador)
    contador -= 1

suma_entradas = 0
while True:
    numero = int(input("Ingrese un número, 0 para terminar: "))
    if numero == 0:
        break
    suma_entradas += numero

print("Suma total de los números ingresados:", suma_entradas)
```

Punto 4. Suma de pares e impares y estado final del ciclo.

```python
suma_pares = 0
suma_impares = 0

for i in range(1, 11):
    if i % 2 == 0:
        suma_pares += i
    else:
        suma_impares += i

estado_final = {
    "suma_pares": suma_pares,
    "suma_impares": suma_impares
}

print("Estado final:", estado_final)
```

El estado cambia dentro del bucle porque en cada iteración las variables acumuladoras suma_pares y suma_impares reciben nuevos valores.

Así se puede ver cómo el programa transforma su memoria paso a paso.

Punto 5. Inventario ingresado por el usuario, validación con continue, producto más caro y estado final.

```python
productos = []

while True:
    nombre = input("Ingrese producto (o 'fin' para salir): ").strip()
    if nombre == "fin":
        break

    precio = float(input("Precio: "))
    cantidad = int(input("Cantidad: "))

    if precio < 0 or cantidad < 0:
        print("Producto omitido por tener precio negativo o cantidad negativa")
        continue

    productos.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})

total = 0
for p in productos:
    total += p["precio"] * p["cantidad"]

print("Total del inventario:", total)

if productos:
    producto_mas_caro = max(productos, key=lambda p: p["precio"])
    print("Producto más caro:", producto_mas_caro)
else:
    print("No se registraron productos válidos.")

print("Estado final del inventario:")
print(productos)
```
