```python

# Teniendo como base el siguiente codigo
# Lista de productos en una tienda
inventario = [
    {"nombre": "Pan", "precio": 1500, "stock": 30},
    {"nombre": "Leche", "precio": 3500, "stock": 15},
    {"nombre": "Café", "precio": 8000, "stock": 10},
]

#Muestra el nombre del primer producto.
print(inventario[0]["nombre"])

#Agrega un nuevo producto al inventario.
nuevo_producto = {"nombre": "Huevos", "precio": 800, "stock": 30}
inventario.append(nuevo_producto)

#Disminuye el stock de “Leche” en 2 unidades (venta).

for p in inventario:
  if p["nombre"] == "leche":
    p["stock"] -= 2

#Calcula el valor total del inventario (precio * stock por producto).
precio_total = 0
for p in inventario:
  precio_total += p["precio"] * p["stock"]
 # Valor del inventario total

#Muestra los productos cuyo stock sea menor a 10.
print("Los productos cuyo stock es menor a 10")
for p in inventario:
  if p["stock"] < 10:
    print(p)

#Aumenta el precio de todos los productos en un 10%.
#Redondea los precios con round(p["precio"], 2) para evitar decimales largos.
for p in inventario:
  p["precio"] = round(p["precio"] * 1.10, 2)


#Busca un producto por nombre (ingresado por el usuario) y muestra el producto encontrado, ten en cuenta que no puede estar.
nombre = input("ingrese el nombre del producto a buscar: ")
encontrado = None
for p in inventario:
  if nombre.lower() == p["nombre"].lower():
    encontrado = p
    break

if encontrado:
  print(f"El producto {p} ha sido encontrado")
else:
  print("El producto no ha sido encontrado")

#Elimina un producto del inventario si su stock es 0.
lista_actualizada = [p for p in inventario if p["stock"] > 0]

#Ordena la lista por precio ascendente.
inventario.sort(key = lambda p: p["precio"])

print("inventario final ordenado")
for p in inventario:
  print(p)
```

Muestra el estado actualizado del sistema (usa print(biblioteca))

```python
biblioteca = {
    "libros": [
        {"titulo": "1984", "autor": "George Orwell", "prestado": False},
        {"titulo": "El Principito", "autor": "Antoine de Saint-Exupéry", "prestado": True},
    ],
    "usuarios": ["Ana", "Luis", "María"]
}

#Muestra cuántos libros están prestados y cuántos disponibles.
prestado = 0
disponibles = 0

for l in biblioteca["libros"]:
  if l["prestado"]:
    prestado += 1
  else:
    disponible += 1
print(f"La cantidada de libros disponibles es {disponibles}")
print(f"La cantidada de libros prestados es {prestado}")

#Agrega un nuevo libro.
nuevo_libro = {"titulo": "1974", "autor": "Messi Ronaldo", "prestado": False}
biblioteca["libros"].append(nuevo_libro)

#Marca como prestado el libro “1984”.
for l in biblioteca["libros"]:
  if l["titulo"] == "1984":
    l["prestado"] = True

#Muestra el estado actualizado del sistema (usa print(biblioteca))
print(biblioteca)
```
