Punto 1. Consumo de la API de usuarios y filtrado por correo que termine en .biz.

```python
import requests

url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
usuarios = response.json()

usuarios_biz = []
for u in usuarios:
    if u["email"].endswith(".biz"):
        usuarios_biz.append(u["name"])

print("Usuarios cuyo correo termina en .biz:")
for nombre in usuarios_biz:
    print("-", nombre)
```

Punto 2. Consulta a la API de Pokémon con condicionales adicionales, listado de habilidades una por una y un uso extra de la respuesta.

```python
import requests

nombre_pokemon = input("Ingrese el nombre de un Pokémon: ").strip().lower()

url = f"https://pokeapi.co/api/v2/pokemon/{nombre_pokemon}"
respuesta = requests.get(url)

if respuesta.status_code == 200:
    datos = respuesta.json()

    pokemon = {
        "nombre": datos["name"],
        "altura": datos["height"],
        "peso": datos["weight"],
        "tipos": [t["type"]["name"] for t in datos["types"]],
        "habilidades": [h["ability"]["name"] for h in datos["abilities"]],
        "stats": {s["stat"]["name"]: s["base_stat"] for s in datos["stats"]}
    }

    print("Información del Pokémon")
    print(f"Nombre: {pokemon['nombre'].capitalize()}")
    print(f"Altura: {pokemon['altura']}")
    print(f"Peso: {pokemon['peso']}")
    print(f"Tipos: {', '.join(pokemon['tipos'])}")
    print(f"Habilidades: {', '.join(pokemon['habilidades'])}")

    if len(pokemon["tipos"]) > 1:
        print("Pokémon de tipo mixto")
    else:
        print("Pokémon de un solo tipo")

    if pokemon["peso"] > 100:
        print("Pokémon de gran tamaño")
    else:
        print("Pokémon ligero")

    print("Habilidades una por una:")
    for habilidad in pokemon["habilidades"]:
        print("-", habilidad)

    total_stats = sum(pokemon["stats"].values())
    stat_mas_alta = max(pokemon["stats"], key=pokemon["stats"].get)

    print("Uso extra de la respuesta de la API")
    print("Suma total de estadísticas base:", total_stats)
    print("Estadística más alta:", stat_mas_alta, "con", pokemon["stats"][stat_mas_alta])

else:
    print("Pokémon no encontrado. Verifique el nombre ingresado.")
```
